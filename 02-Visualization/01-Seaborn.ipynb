### import

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### read 'heart.csv'  under DATA folder

```python
df = pd.read_csv('../DATA/heart.csv')
```

```python
df.head()
```

### create a distribution plot of 'age': kde=hide, bins=30, color=red, plot size= 8*4 inch.

```python
plt.figure(figsize=(8,4))
sns.distplot(df['age'],kde=False,bins=30,color='red');
```

### create a count plot of 'cp' (cheast pain) per sex: palette= terrain

```python
sns.countplot(x='cp',data=df,hue='sex',palette='terrain')
```

### create a box plot of heart rate 'thalach' per target 'target' per sex

```python
sns.boxplot(x='target',y='thalach',data=df,hue='sex')
```

### create a scatter plot between cholestrol ('chol') vs heartrate ('trestbps') split by sex: palette='Dark2', point size= based on age

```python
sns.scatterplot(x='chol',y='trestbps',data=df,hue='sex',palette='Dark2',size='age')
```
