```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
diamonds = pd.read_csv('../DATA/diamonds.csv')
```

```python
diamonds.head()
```

```python
diamonds.info()
```

### Create a scatterplot of price versus carat 

```python
sns.scatterplot(x='carat',y='price',data=diamonds)
```

### Use alpha parameter and edgecolor parameter to deal with the overlapping issue and white edgemarker issue.

```python
sns.scatterplot(x='carat',y='price',data=diamonds, alpha=0.1,edgecolor='#2375a1')
```

### Make the previous plot larger

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='carat',y='price',data=diamonds, alpha=0.1,edgecolor='#2375a1')
```

### Create a histogram of the price column

```python
plt.figure(figsize=(12,8))
sns.distplot(diamonds['price'],kde=False);
```

### Create a count plot of the instances per cut type

```python
sns.countplot(x='cut',data=diamonds)
```

### Create a large box plot figure showing the price distribution per cut type

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='cut',y='price',data=diamonds);
```

### change the following to reflect this relationship between the cut types: the ordering of the boxplot, coloring.

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='cut',y='price',data=diamonds,order=['Fair','Ideal','Good','Very Good','Premium'],palette='cool');
```
